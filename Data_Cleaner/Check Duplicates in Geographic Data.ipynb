## Quick Removal of Duplicates from Locations
* Remove duplicates of country_code/city for this data
* Ensures that there is only one value for each country_code/city pairing
* Addendum to "Scraping Coordinates" notebook, but it was getting messy

```python
current_df = pd.read_csv("geographical_data/coord_locations.csv")
current_df
```

```python
final_df = current_df.drop_duplicates(subset = ["country_code", "city"], ignore_index = True)
final_df
```

```python
#output updated csv
final_df.to_csv("geographical_data/coord_locations.csv", index=False)
```

```python
#output updated json
final_df.to_json("./geographical_data/coord_locations.json", orient="index")
```
