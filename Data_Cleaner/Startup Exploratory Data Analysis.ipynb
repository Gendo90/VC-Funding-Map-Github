```python
import pandas as pd
```

```python
output_df = pd.read_csv("../VC_Data/Dirty_Data/investments_VC.csv", encoding="ISO-8859-1")
output_df.head(20)
```

```python
#might want to keep some missing fields if they can be filled in later by scraping/crunchbase API's
#drop "state_code" column for dataframe - no cities outside US and Canada have state values
cleaned_df = output_df.drop(columns=["state_code"])
cleaned_df
```

```python
#remove rows with all null values
cleaned_df = cleaned_df.dropna(how="all")
cleaned_df
```

```python
#remove rows missing BOTH a country_code and city values - cannot get geographical data in this case
#can still use startup data for country data visualization even if city is missing
cleaned_df = cleaned_df.dropna(subset=["country_code", "city"], thresh=2)
cleaned_df
```

```python
#remove leading and trailing spaces from columns
cleaned_df.columns = cleaned_df.columns.map(lambda x: x.strip())
print(cleaned_df.columns)
```

```python
#remove rows where total funding is nonexistent (given as a null)
cleaned_df = cleaned_df.dropna(subset=["funding_total_usd"])
cleaned_df
```

```python
#remove rows where total funding is given as "-"
cleaned_df["funding_total_usd"] = cleaned_df["funding_total_usd"].apply(lambda x: x.strip())
cleaned_df = cleaned_df.loc[cleaned_df["funding_total_usd"] != "-", :]
cleaned_df
```

```python
#change total funding format to a number instead of a string (in USD)
cleaned_df["funding_total_usd"] = cleaned_df["funding_total_usd"].apply(lambda x: int("".join([a for a in x if a != ","])))
cleaned_df
```

```python
#remove values where the "last_funding_at" was earlier than 2005
#we are only looking at the date range of 2005 through 2015

#first, convert to datetime column
cleaned_df["last_funding_at"] = pd.to_datetime(cleaned_df["last_funding_at"])
cleaned_df = cleaned_df.loc["2004-12-31" < cleaned_df["last_funding_at"]]
cleaned_df
```
